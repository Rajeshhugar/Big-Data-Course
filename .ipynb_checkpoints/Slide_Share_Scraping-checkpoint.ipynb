```python
import requests
import os
from pathlib import Path
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
import time

def download_slides():
    # Create folders if they don't exist
    base_folder = Path("hadoop_slides")
    images_folder = base_folder / "images"
    images_folder.mkdir(parents=True, exist_ok=True)
    
    # Base URL pattern
    base_url = "https://image.slidesharecdn.com/2-210731071428/75/Introduction-to-Hadoop-{}-2048.jpg"
    
    # Download images
    downloaded_files = []
    for slide_num in range(1, 30):  # 1 to 29
        url = base_url.format(slide_num)
        output_file = images_folder / f"slide_{slide_num:02d}.jpg"
        
        try:
            response = requests.get(url, timeout=10)
            if response.status_code == 200:
                with open(output_file, 'wb') as f:
                    f.write(response.content)
                downloaded_files.append(output_file)
                print(f"Downloaded slide {slide_num}")
            else:
                print(f"Failed to download slide {slide_num}: Status code {response.status_code}")
                
            # Be nice to the server
            time.sleep(1)
            
        except Exception as e:
            print(f"Error downloading slide {slide_num}: {str(e)}")
    
    return downloaded_files

def create_pdf(image_files):
    # Create PDF
    pdf_path = Path("hadoop_slides/hadoop_presentation.pdf")
    c = canvas.Canvas(str(pdf_path), pagesize=letter)
    
    for img_file in image_files:
        try:
            # Add image to PDF
            c.drawImage(str(img_file), 50, 50, width=500, height=400)
            c.showPage()
        except Exception as e:
            print(f"Error adding {img_file} to PDF: {str(e)}")
    
    c.save()
    print(f"PDF created at {pdf_path}")

def main():
    print("Starting download of slides...")
    image_files = download_slides()
    
    if image_files:
        print("\nCreating PDF...")
        create_pdf(image_files)
        print("\nProcess completed!")
    else:
        print("\nNo images were downloaded. Cannot create PDF.")

if __name__ == "__main__":
    main()
```
